# The Sinus Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

from mycelium.activation import Sin
from mycelium.basic_units import radians, degrees
```

The ...


```python
low, high = -np.pi * 2, np.pi * 2
x_vals = np.linspace(low, high, 100)
x = np.array([val*radians for val in x_vals])
function = Sin()
y = function(x_vals)

plt.axhline(y=0, lw=1, color='dimgray')
plt.axvline(x=0, lw=1, color='dimgray')

plt.plot(x, y, label=function.latex(), xunits=radians)
plt.grid(True)
plt.legend(fontsize=16)
plt.title("Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

```python
plt.plot(x, y, label=function.latex(), xunits=degrees)
plt.axhline(y=0, lw=1, color='dimgray')
plt.axvline(x=0, lw=1, color='dimgray')
plt.grid(True)
plt.legend(fontsize=16)
plt.title("Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
