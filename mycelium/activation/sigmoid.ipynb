# The Sigmoid Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

from mycelium.activation import Sigmoid
```

The ...


```python
x = np.linspace(-8, 8, 100)
function = Sigmoid()
y = function(x)

plt.axhline(y=0, lw=1, color='dimgray')
plt.axvline(x=0, lw=1, color='dimgray')

plt.plot(x, y, label=function.latex())
plt.grid(True)
plt.legend(fontsize=16)
plt.title("Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
