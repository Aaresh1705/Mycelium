# The Linear Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

from mycelium.activation import Linear
```

The linear activation function just outputs the input. It is characterized by the following properties:
- **From:** $f(x) = x$.
- **Linearity:** it is a linear function with slope 1 and intercept 0.
- **No nonlinearity introduced:** It does not transform the input, since the output equals the input.
- **Derivative:** $f'(x) = 1$ for all $x$.

Because it introduces no nonlinearity, using this activation in hidden layers causes the network to collapse into an equivalent single linear transformation, regardless of depth. For this reason, it is typically used only in output layers, especially for regression tasks where unbounded continuous outputs are required.


```python
x = np.linspace(-1, 1, 100)
function = Linear()
y = function(x)

plt.axhline(y=0, lw=1, color='dimgray')
plt.axvline(x=0, lw=1, color='dimgray')

plt.plot(x, y, label=function.latex())
plt.grid(True)
plt.legend(fontsize=16)
plt.title("Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
